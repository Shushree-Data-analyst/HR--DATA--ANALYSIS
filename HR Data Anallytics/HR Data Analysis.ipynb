```python
# Calculate total attrition (where attrition is 'Yes')
attrition_count = len(hr[hr['attrition'] == 'Yes'])

# Calculate total employees
total_employees = len(hr)

# Calculate rate
attrition_rate = (attrition_count / total_employees) * 100

print(f"Attrition Rate: {round(attrition_rate, 2)}%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.countplot(data=hr[hr['attrition'] == 'Yes'], 
              x='education_field', 
              palette='magma',
              order=hr[hr['attrition'] == 'Yes']['education_field'].value_counts().index)

plt.title('Attrition Count by Education Field')
plt.xticks(rotation=45)
plt.show()
```

```python
# 1. Filter the data for employees who left
attrition_yes = hr[hr['attrition'] == 'Yes']

# 2. Get the counts per department
dept_counts = attrition_yes['department'].value_counts()

# 3. Create the Pie Plot
plt.figure(figsize=(8, 8))
plt.pie(dept_counts, 
        labels=dept_counts.index, 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=sns.color_palette('Set2')) # Fixed: Changed 'set2' to 'Set2'

plt.title('Percentage of Attrition by Department')
plt.show()
```

```python
# 1. Define the age bins and labels
bins = [18, 25, 35, 45, 55, 60]
labels = ['18-24', '25-34', '35-44', '45-54', '55+']

# 2. Create a new Age Group column
hr['age_group'] = pd.cut(hr['age'], bins=bins, labels=labels, right=False)

# 3. Create the Bar Plot
plt.figure(figsize=(10, 6))
sns.countplot(data=hr, x='age_group', palette='Blues_d')

plt.title('Number of Employees by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Employee Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Boxplot comparing Age across Marital Status, split by Attrition
sns.boxplot(data=hr, x='marital_status', y='age', hue='attrition', palette='Set1')

plt.title('Age Distribution by Marital Status and Attrition')
plt.xlabel('Marital Status')
plt.ylabel('Age')
plt.legend(title='Attrition', loc='upper right')
plt.show()
```
